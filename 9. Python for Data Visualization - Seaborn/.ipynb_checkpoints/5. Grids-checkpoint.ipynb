```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create a PairGrid
g = sns.PairGrid(iris)

# Map diagonal plots (histogram)
g.map_diag(sns.histplot)

# Map upper plots (scatter)
g.map_upper(plt.scatter)

# Map lower plots (kernel density estimate)
g.map_lower(sns.kdeplot)

# Show the plot
plt.show()
```
