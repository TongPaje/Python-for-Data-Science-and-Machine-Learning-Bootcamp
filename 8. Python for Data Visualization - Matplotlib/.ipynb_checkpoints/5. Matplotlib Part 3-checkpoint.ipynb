```python
import matplotlib.pyplot as plt


# Define x and y data
x = [0, 0.2, 0.4, 0.6, 0.8, 1]
y = [0, 0.4, 0.7, 1.2, 1.5, 2]

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='green', lw=2, linestyle='--', marker='o', markersize=10, 
       markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='red')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
